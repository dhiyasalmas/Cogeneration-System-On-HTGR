Exergi adalah nilai kerja teoritis yang dapat dibangkitkan dari suatu sistem saat mengalami kesetimbangan dengan lingkungan sekitarnya. Secara sederhana **exergi digunakan untuk melihat derajat kemanfaatan suatu sistem secara keseluruhan**.

**Exergi total** dapat dihitung dengan rumus :
$$ E = E^{PH} + E^{KN} + E^{PT} + E^{CH} $$

Dimana exergi terdiri dari beberapa komponen, seperti :
1. **Exergi fisik $E^{PH}$**, Perbedaan energi antara keadaan sistem dan lingkungan dengan menggunakan hukum termodinamika. 
$$ E^{PH} = (h - h_0) - T_0 (s - s_0) $$
dengan:
- $h$ : entalpi sistem (kJ/kg)
- $h_0$ : entalpi lingkungan (kJ/kg)
- $T_0$ : suhu lingkungan (K)
- $s$ : entropi sistem (kJ/kg.K)
- $s_0$ : entropi lingkungan (kJ/kg.K)

2. **Exergi kinetik $E^{KN}$**, Energi akibat kecepatan sistem.
$$E^{KN} = \frac{1}{2} m v^2 $$

3. **Exergi potensial $E^{PT}$**, Energi akibat posisi sistem dalam medan gravitasi.
$$E^{PT} =  m g h $$

4. **Exergi kimia $E^{CH}$**, Potensi energi yang berasal dari reaksi kimia dalam sistem. Untuk suatu zat murni exergi kimia :
$$ E^{CH} = \Sigma_i n_i \mu_i^0 = \Sigma_i n_i \mu_i $$
dengan:
- $n_i$ : jumlah mol komponen i
- $\mu_i^0$ : potensial kimia standar komponen i
- $\mu_i$ : potensial kimia komponen i


Untuk menghitung **exergi fisik dibutuhkan data entalpi dan entropi**. **Entalpi (H)** adalah **jumlah energi yang tersimpan dalam suatu sistem**, sedangkan **entropi (S)** adalah **ketidakberaturan sistem**.

Namun dalam perhitungan **entalpi dan entropi perlu di tinjau keadaan panas**, dimana **panas terdiri atas 2 jenis perpindahan**.
- **Panas sensibel**, dimana hanya **mengubah suhu** tanpa mengubah fase zat.
- **Panas laten**, dimana **mengubah fase zat** tanpa mengubah suhu.

Dengan menggunakan **Hukum Hess** maka **entalpi** dapat di hitung dengan :
$$ \Delta H_f = \sum \Delta H_{f produk} - \sum \Delta H_{f reaktan} $$
dimana untuk **panas sensibel** $\Sigma \Delta H_{f produk}$ dan $\Sigma \Delta H_{f reaktan}$ dapat dihitung dengan : 
$$ \sum \Delta H_{f produk} =  m C_p \Delta T $$
dengan
- $C_p$ : kalor jenis zat (J/kg.K atau cal/g.C)

sedangkan untuk **panas laten**
$$ \sum \Delta H_{f produk} = m L $$
dengan
- L : kalor laten (J/kg atau cal/g)

Untuk **entropi** menggunakan **Hukum kedua termodinamika** dimana :
$$ \Delta S_{universe} = \Delta S_{sistem} + \Delta S_{lingkungan} $$
dimana dalam **entropi sistem biasanya terdapat reaksi antara reaktan dan produk** yang dapat dihitung dengan persamaan:
$$ \Delta S_{reaksi} = \sum \Delta S_{produk} - \sum \Delta S_{reaktan} $$
nilai $\Sigma \Delta S_{produk}$ dan $\Sigma \Delta S_{reaktan}$ dalam **panas sensibel**, dapat dihitung dengan :
$$ \Delta S_{produk} = m C_p ln (\frac{T_2}{T_1})$$
untuk **panas laten**
$$ \Delta S_{produk} = \frac{m L}{T}$$

Catatan:
> Pada kondisi lingkungan maka di gunakan data temperatur $25^{\circ} C$ dan tekanan 1 atm.

Dimana kalor jenis ($C_p$) untuk entalpi dan entropi akan memiliki nilai yang berbeda beda tergantung pada suhu, hal ini dapat dihitung dengan menggunakan rumus:
$$ \int_{T_0}^{T} \frac{\Delta C_p^0}{R} dT = \Delta A (T-T_0) + \frac{\Delta B}{2} (T^2 - T_0^2) + \frac{\Delta C}{3} (T^3 - T_0^3) + \Delta D(\frac{T-T_0}{TT_0}) $$

nilai A, B, C dan D merupakan konstanta yang didapatkan dari eksperimen dan dapat ditemukan dalam buku termodinamika atau kitab suci teknik kimia (Perry Chemical Engineering Handbook), seperti berikut.
